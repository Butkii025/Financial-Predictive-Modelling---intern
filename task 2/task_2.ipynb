```python
df = pd.read_csv('SQL_Sales_Dataset_200_Rows.csv')

print("--- FIRST 5 ROWS ---")
print(df.head())

print("\n--- BASIC INFORMATION ---")
df.info()

print("\n--- SHAPE OF THE DATASET ---")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

## 2. Handle missing values and duplicates using Pandas

```python
print("--- Missing Values Per Column ---")
print(df.isnull().sum())

duplicate_count = df.duplicated().sum()
print(f"\n--- Total Duplicate Rows Found: {duplicate_count} ---")
```

```python
df.drop_duplicates(inplace=True)
print("Duplicates removed successfully!")
```

```python
df.dropna(inplace=True)
```

## 3: Group Data by Category and Find Total Revenue

```python
category_revenue = df.groupby('category')['total_price'].sum().reset_index()

category_revenue.columns = ['Product Category', 'Total Revenue']

category_revenue = category_revenue.sort_values(by='Total Revenue', ascending=False)

print("--- TOTAL REVENUE BY CATEGORY ---")
print(category_revenue)
```

## 4: Sort Data by Multiple Columns Using Python

```python
sorted_df = df.sort_values(by=['category', 'total_price'], ascending=[True, False])

print("--- SORTED DATASET (BY CATEGORY & TOTAL PRICE) ---")
print(sorted_df[['category', 'product_name', 'total_price']].head(10))
```

## 5: Create a Correlation Matrix for Numerical Columns

```python
numerical_df = df.select_dtypes(include=['number'])

correlation_matrix = numerical_df.corr()

print("--- CORRELATION MATRIX ---")
print(correlation_matrix)
```

```python
# Save the sorted, cleaned dataframe to a new CSV file
sorted_df.to_csv('SQL_Sales_Dataset_Processed.csv', index=False)

print("Success! Your final output file 'SQL_Sales_Dataset_Processed.csv' has been saved to your workspace directory.")
```
